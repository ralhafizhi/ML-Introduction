```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes() #koordinatnya
ax.plot(x,y)
```

```python
#plot dua sekaligus dalam satu figure
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
```

### Shorcut
--->Jalan pintas lebih singkat

```python
plt.figure(figsize=(6,6))
plt.plot(x,y1, 'r-', linewidth=5)
plt.plot(x,y2, 'bo', markersize=5)

#menambahkan limit pada x dan y
plt.xlim(0,10)
plt.ylim(-2,2)

#merubah axis agar jadi balance antara x dan y (skalanya sama)
plt.axis('equal')

#menambahkan judul
plt.title('Sin dan Cos')
#menambahkan label
plt.xlabel('x')
plt.ylabel('y')
```

```python
fig=plt.figure(figsize=(6,6))
ax = plt.axes() #koordinatnya

ax.plot(x,y1, 'r-', linewidth=5)
ax.plot(x,y2, 'bo', markersize=5)

#menambahkan limit pada x dan y
ax.set_xlim(0,10)
ax.set_ylim(-2,2)

#menambahkan judul
ax.set_title('Sin dan Cos', fontsize=20)

#menambahkan label
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)

#merubah axis agar jadi balance antara x dan y (skalanya sama)
ax.axis('equal');
```

# Menambahkan Legend

```python
fig=plt.figure(figsize=(6,6))
ax = plt.axes() #koordinatnya

ax.plot(x,y1, 'r-', linewidth=5, label='sin')
ax.plot(x,y2, 'bo', markersize=5, label='cos')

#menambahkan limit pada x dan y
ax.set_xlim(0,10)
ax.set_ylim(-2,2)

#menambahkan judul
ax.set_title('Sin dan Cos', fontsize=20)

#menambahkan label
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)

#merubah axis agar jadi balance antara x dan y (skalanya sama)
ax.axis('equal');

#menambahkan legend
ax.legend(loc='lower center', ncol=20, shadow=True, borderpad=1);
```

# Scatter Plot

```python
df = pd.read_csv('data/circle_data.csv', delimiter=';')
df.head()
```

```python
#memisahkan x dan y dari data frame diatas
x = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(x,y)
plt.axis('equal');
```

```python
#menambahkan warna,size dan atribut lainnya
plt.scatter(x,y, c='r', s=20)
plt.axis('equal');
```

```python
#merubah masing-masing elemen berdasarkan radiusnya
plt.scatter(x,y, c=rad, s=rad*90, cmap='seismic')
plt.axis('equal');
```
