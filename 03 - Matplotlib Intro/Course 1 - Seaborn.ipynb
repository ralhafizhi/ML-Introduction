# Kombinasi Seaborn dan Data Frame (Pandas)

## Statistical Summary

```python
df = pd.read_csv('data/cereal.csv', index_col='name')
df.head()
```

```python
df.shape
```

## Plot Missing Value

```python
#fungsi isna() untuk melihat missing value
plt.figure(figsize=(15,15))
sns.heatmap(df.isna(), cmap='Blues');
```

## Plot korelasi antar Kolom

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), cmap='twilight_shifted', cbar=True, vmin=-1, vmax=1, annot=True, square=True, fmt='.1f')
```

## Pairplot

```python
sns.pairplot(df)
```

```python
#pairplot dengan beberapa kolom tertentu, berdasarkan shelf gunanya untuk melihat tren antar variabel
sns.pairplot(df, hue='shelf', vars=['calories','fat','rating'])
```

```python
#pairplot dengan beberapa kolom tertentu, berdasarkan shelf gunanya untuk melihat tren antar variabel
sns.pairplot(df, hue='shelf', vars=['calories','fat','rating'], kind='reg');
```

## JoinPlot

```python
sns.jointplot('potass','fiber', data=df, marginal_kws={'bins': 25}, kind='reg')
```

## Categorical Data

## Persebaran

```python
sns.catplot('mfr','rating', data=df)
```

```python
#kuartil 1,2 dan 3
sns.boxplot('mfr', 'rating', data=df);
```

```python
plt.figure(figsize=(7,7))
sns.violinplot('mfr','rating', data=df);
```

## Count

```python
sns.catplot('mfr', data=df, kind='count');
```

```python
#bedanya adalah ini pakai countplot dan bisa menampilkan hue(berdasrkan kolom apa)
sns.countplot('mfr', data=df, hue='shelf');
```
